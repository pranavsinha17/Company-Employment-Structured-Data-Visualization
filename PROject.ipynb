```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\\Machine Learning\\ML\\google.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('dark_background')
df['Company'].value_counts().plot.bar(color = 'lightpink')
plt.title("Company Types", color = 'black')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('fivethirtyeight')
df['Title'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightgreen')
plt.title("Top 20 Titles", color = 'black')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('fivethirtyeight')
df['Category'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightgreen')
plt.title("Top 20 category", color = 'black')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('fivethirtyeight')
df['Location'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'pink')
plt.title("Top 20 Analysis", color = 'black')
plt.show()
```

```python
from wordcloud import WordCloud
plt.rcParams['figure.figsize'] = (15, 15)
wc = WordCloud(background_color = 'yellow', max_words = 80).generate(str(df['Responsibilities']))
plt.title('Buzz Responsibilities', color = 'black', fontsize = 20)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
